Plotting the Unemployment Rate over time for a specific region and area


```python
selected_region = 'Andhra Pradesh'
selected_area = 'Rural'
```

```python

filtered_df = data[(data['Region'] == selected_region) & (data['Area'] == selected_area)]

plt.figure(figsize=(12, 6))
plt.plot(filtered_df[' Date'], filtered_df[' Estimated Unemployment Rate (%)'], linestyle='-', marker='o')
plt.title(f'Unemployment Rate Over Time in {selected_region} ({selected_area})')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

Regional Analysis:

Explore how the unemployment rate varies across different regions.

```python

import seaborn as sns

# Box plot of unemployment rate by region
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=data)
plt.title('Unemployment Rate Across Regions')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Area-wise Analysis:


Compare unemployment rates between rural and urban areas.

```python
# Box plot of unemployment rate by area
plt.figure(figsize=(8, 6))
sns.boxplot(x='Area', y=' Estimated Unemployment Rate (%)', data=data)
plt.title('Unemployment Rate Across Areas')
plt.show()
```
