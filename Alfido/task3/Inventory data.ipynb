# Create a histogram of product quantities

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=inventory_data, x='Quantity', bins=20)
plt.title('Distribution of Product Quantities')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

# Create a scatter plot of product prices (Dollars) vs. quantities

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=inventory_data, x='Dollars', y='Quantity')
plt.title('Price vs. Quantity')
plt.xlabel('Price (Dollars)')
plt.ylabel('Quantity')
plt.show()
```

# Create a bar plot to visualize the counts of products by VendorName

```python
plt.figure(figsize=(12, 6))
vendor_counts = inventory_data['VendorName'].value_counts().iloc[:100]
sns.barplot(x=vendor_counts.index, y=vendor_counts.values, palette="viridis")
plt.title('Product Count by Vendor')
plt.xlabel('Vendor Name')
plt.ylabel('Product Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='VendorName', y='Quantity', data=inventory_data.head(20))
plt.title('Inventory by Vendor')
plt.xlabel('Vendor Name')
plt.ylabel('Quantity')
plt.xticks(rotation=45)
plt.show()
```
